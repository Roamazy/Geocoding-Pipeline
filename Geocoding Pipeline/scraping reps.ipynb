```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.palegis.us/house/members'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

members = []

for member in soup.find_all('div', class_='member'):
    name = member.get('data-name')
    county = member.get('data-county')
    party = member.get('data-party')
    district = member.get('data-district')
    link_tag = member.find('a', class_='thumb-info-wrapper')
    link = f"https://www.palegis.us{link_tag['href'].strip()}" if link_tag else None
    
    members.append({
        'Name': name,
        'County': county,
        'Party': party,
        'District': district,
        'Profile Link': link
    })

for m in members:
    print(m)
```

```python
house_reps = pd.DataFrame(members)
house_reps
```

```python
house_reps["District"] = house_reps["District"].astype(str).str.lstrip("0")
house_reps
```

```python
house_reps.drop(columns=["County", "Profile Link"], inplace=True)
house_reps
```
